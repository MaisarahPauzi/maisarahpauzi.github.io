## Linear Regression from Scratch 
Create Fitted Line (Prediction)


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('income.data.csv')
df.head()
```

```python
X = df['income']
Y = df['happiness']
```

```python
plt.scatter(X,Y)
plt.plot()
```

**Linear Regression Equation**




```python
# find mean of X & Y

mean_X = X.mean()
mean_Y = Y.mean()

# find B1
numerator = np.sum((X-mean_X)*(Y-mean_Y))
denominator = np.sum((X-mean_X)**2)

b1 = numerator/denominator
b1
```

```python
# find b0
b0 = mean_Y - (b1*mean_X)
b0
```

```python
# Predictions array

y_hat = b0 + b1*X
```

```python
plt.plot(X, y_hat, c="red")
```

```python
plt.scatter(X,Y)
plt.plot(X, y_hat, c="red")
```

**R-squared Metric**

Measures the goodness of fit of the linear regression model
 - SSE is the sum of squared errors of the regression model
 - SST is the sum of errors of the baseline model
 - The baseline model always predicts mean of Y






```python
SSE = np.sum((df['happiness']-y_hat)**2)
SST = np.sum((df['happiness']-mean_Y)**2)

R2 = 1 - SSE/SST
R2
```
