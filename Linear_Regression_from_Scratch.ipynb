Create Fitted Line (Prediction)


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('income.data.csv')
df.head()
```

```python
plt.scatter(df['income'], df['happiness'])
plt.plot()
```

**Linear Regression Equation**

image.png

```python
# find mean of X & Y

mean_X = df['income'].mean()
mean_Y = df['happiness'].mean()

# find B1
numerator = np.sum((df['income']-mean_X)*(df['happiness']-mean_Y))
denominator = np.sum((df['income']-mean_X)**2)

b1 = numerator/denominator
b1
```

```python
# find b0
b0 = mean_Y - (b1*mean_X)
b0
```

```python
# Predictions array

y_hat = b0 + b1*df['income']
```

```python
plt.plot(df['income'], y_hat, c="red")
```

```python
plt.scatter(df['income'], df['happiness'])
plt.plot(df['income'], y_hat, c="red")
```
